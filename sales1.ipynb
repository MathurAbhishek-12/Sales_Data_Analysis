### Q4. What products are most often sold together?

```python
all_data.head()
```

```python
all_data.info()
```

```python
df = all_data[all_data["Order ID"].duplicated(keep=False)]
df.head()
```

''' keep = {‘first’, ‘last’, False}, default ‘first’
                  Determines which duplicates (if any) to mark.

                  first : Mark duplicates as True except for the first occurrence.

                  last : Mark duplicates as True except for the last occurrence.
                    
                  False : Mark all duplicates as True.'''

```python
df["Grouped"] = df.groupby(["Order ID"])["Product"].transform(lambda x: ",".join(x))  #earlier we used;   all_data.groupby("Month").sum()
df.head()
```

```python
df = df[["Order ID", "Grouped"]]
df.head()
```

```python
df = df.drop_duplicates()
df
```

```python
# Referred: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary

from itertools import combinations
from collections import Counter

count = Counter()

for row in df["Grouped"]:
    count.update(Counter(combinations(row, 2)))

print(count)
```

```python
# As we donot want combination of any 2 digit in a row, we want the combination of 2 words,therefore:
from itertools import combinations
from collections import Counter

count = Counter()

for row in df["Grouped"]:
    row_list = row.split(",")
    count.update(Counter(combinations(row_list, 2)))   # Try 3

print(count)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df["Grouped"]:
    row_list = row.split(",")
    count.update(Counter(combinations(row_list, 2))) 
    
for i,j in count.most_common(15):   #  Or directly use:  count.most_common(10)
    print (i,j)
```
